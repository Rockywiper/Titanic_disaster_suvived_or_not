```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
train = pd.read_csv(r'C:\Users\91938\Desktop\Machine learning\train.csv')
```

```python
test = pd.read_csv(r'C:\Users\91938\Desktop\Machine learning\test.csv')
```

```python
data = pd.concat([train, test])
```

```python
data['Age'].fillna('0', inplace = True)
```

```python
data['Fare'].fillna('0', inplace = True)
```

```python
data['Cabin'].fillna('unknown', inplace = True)
```

```python
one_hot = pd.get_dummies(data, columns = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked'])
```

```python
one_hot.head()
```

```python
one_hot.drop(columns = ['Name', 'Ticket', 'Fare', 'Cabin'], inplace = True)
```

```python
one_hot = one_hot.reset_index(drop = True)
```

```python
data_preprocessed = pd.DataFrame(one_hot)
```

```python
data_preprocessed.head(10)
```

```python
data_preprocessed['Age'].unique()
```

```python
preprocessed_train = data_preprocessed[data_preprocessed['Survived'].notna()]
preprocessed_test = data_preprocessed[data_preprocessed['Survived'].isna()]
```

```python
x = preprocessed_train.drop(columns = ['PassengerId', 'Survived'])
y = preprocessed_train['Survived']
```

### Train_test_split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 40)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Fitting the model

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

### Predicting train set

```python
y_pred_train = model.predict(x_train)
```

```python
print('accuracy_score:',accuracy_score(y_train, y_pred_train))
print('f1_score:', f1_score(y_train, y_pred_train))
print('confusion_matrix:\n', confusion_matrix(y_train, y_pred_train))
```

### Predicting test set

```python
y_pred_test = model.predict(x_test)
```

```python
print('accuracy_score:',accuracy_score(y_test, y_pred_test))
print('f1_score:', f1_score(y_test, y_pred_test))
print('confusion_matrix:\n', confusion_matrix(y_test, y_pred_test))
```

```python
preprocessed_x_test = preprocessed_test.drop(columns=['PassengerId','Survived'])
y_prediction = model.predict(preprocessed_x_test)
```

```python
sample['Survived'] = y_prediction
sample.to_csv('submission_titanic_disaster_prediction.csv',index=False)
```
